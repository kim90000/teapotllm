```python
from teapotai import TeapotAI
# Sample documents (in practice, these could be articles or longer documents)
documents = [
    "The Eiffel Tower is located in Paris, France. It was built in 1889 and stands 330 meters tall.",
    "The Great Wall of China is a historic fortification that stretches over 13,000 miles.",
    "The Amazon Rainforest is the largest tropical rainforest in the world, covering over 5.5 million square kilometers.",
    "The Grand Canyon is a natural landmark located in Arizona, USA, carved by the Colorado River.",
    "Mount Everest is the tallest mountain on Earth, located in the Himalayas along the border between Nepal and China.",
    "The Colosseum in Rome, Italy, is an ancient amphitheater known for its gladiator battles.",
    "The Sahara Desert is the largest hot desert in the world, located in North Africa.",
    "The Nile River is the longest river in the world, flowing through northeastern Africa.",
    "The Empire State Building is an iconic skyscraper in New York City that was completed in 1931 and stands at 1454 feet tall."
]


# Initialize TeapotAI with documents for RAG
teapot_ai = TeapotAI(documents=documents)

# Get the answer using RAG
answer = teapot_ai.chat([
    {
        "role":"system",
        "content": "You are an agent designed to answer facts about famous landmarks."
    },
    {
        "role":"user",
        "content": "What landmark was constructed in the 1800s?"
    }
])
print(answer) # => The Eiffel Tower was constructed in the 1800s.
```

```python
import fitz  # PyMuPDF
from teapotai import TeapotAI
from transformers import AutoTokenizer

# Load and extract text from PDF
pdf_path = '/content/Understanding_Climate_Change.pdf'
doc = fitz.open(pdf_path)
pdf_text = ""

# Extract text from all pages
for page in doc:
    pdf_text += page.get_text()

doc.close()

# Initialize Tokenizer (use the same tokenizer as TeapotAI)
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-mpnet-base-v2")  # Or the one used by TeapotAI

# Function to chunk text into smaller segments
def chunk_text(text, tokenizer, max_length=512, overlap=50):
    """
    Chunks the input text into segments with a maximum length.

    Args:
        text: The input text to chunk.
        tokenizer: The tokenizer to use for token counting.
        max_length: The maximum length of each chunk (in tokens).
        overlap: The amount of overlap between consecutive chunks (in tokens).  Helps preserve context.

    Returns:
        A list of text chunks.
    """
    sentences = text.split(". ")  # split into sentences

    chunks = []
    current_chunk = ""
    current_length = 0

    for sentence in sentences:
        sentence_length = len(tokenizer.tokenize(sentence))

        if current_length + sentence_length <= max_length:
            current_chunk += sentence + ". "
            current_length += sentence_length + 1 # add 1 for the period and space.
        else:
            chunks.append(current_chunk)
            # Start a new chunk with overlap
            overlap_tokens = tokenizer.tokenize(current_chunk)[-overlap:]
            overlap_text = tokenizer.decode(tokenizer.convert_tokens_to_ids(overlap_tokens))

            current_chunk = overlap_text + sentence + ". "  # Add overlapping part
            current_length = len(tokenizer.tokenize(current_chunk)) # Recalculate length after overlap

    # Add the last chunk if it's not empty
    if current_chunk:
        chunks.append(current_chunk)

    return chunks

# Chunk the PDF text
chunks = chunk_text(pdf_text, tokenizer)

# Initialize TeapotAI with the chunks as documents
teapot_ai = TeapotAI(documents=chunks)

# Ask question
query = "What is the topic of the book?"
answer = teapot_ai.query(query=query)  # No need to pass the whole PDF text as context now.  The document is already indexed.

print(answer)
```

```python
import pandas as pd
from teapotai import TeapotAI

# تحميل ملف CSV
csv_path = '/content/synthqa.csv'
df = pd.read_csv(csv_path)

# تهيئة TeapotAI بدون تحميل المستندات مسبقًا
teapot_ai = TeapotAI()

# تحديد الاستعلام الخاص بك
search_term = "Galactic Odyssey (2020)"

# البحث في كل صف بشكل منفصل
results = []
for i, row in df.iterrows():
    # أخذ حقل السؤال فقط
    if 'question' in df.columns:
        question_text = row['question']

        # تخطي إذا كان النص طويلاً جدًا
        if len(question_text.split()) > 400:  # تقدير تقريبي للتوكنات
            continue

        # معالجة الصفوف التي قد تحتوي على معلومات ذات صلة فقط
        if search_term.lower() in question_text.lower():
            try:
                result = teapot_ai.query(query=search_term, context=question_text)
                results.append(f"صف {i}: {result}")
                print(f"تمت معالجة الصف {i}")  # للتأكد من أن الكود يعمل
            except Exception as e:
                print(f"خطأ في معالجة الصف {i}: {e}")

    # اختياري: إضافة حد لعدم معالجة الكثير من الصفوف
    if i > 50:  # معالجة أول 50 صف فقط
        break

# طباعة النتائج
print("\n--- النتائج ---")
for result in results:
    print(result)
```

```python
import pandas as pd
from teapotai import TeapotAI
from transformers import AutoTokenizer

# Load CSV file into a Pandas DataFrame
csv_path = '/content/synthqa.csv'
df = pd.read_csv(csv_path)

# Access the text content from the correct column
# Assuming the text is in a column named 'question'
questions = df['question'].tolist()

# Initialize TeapotAI without pre-loading documents
teapot_ai = TeapotAI()

# Initialize Tokenizer
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-mpnet-base-v2")

def chunk_text(text, max_length=500):
    """
    Chunks the input text into segments with a maximum length.

    Args:
        text: The input text to chunk.
        max_length: The maximum length of each chunk (in tokens).

    Returns:
        A list of text chunks.
    """
    # Get token IDs directly
    token_ids = tokenizer(text, return_tensors="pt").input_ids[0].tolist()

    # Split token IDs into chunks
    num_chunks = len(token_ids) // max_length + 1
    chunks = [token_ids[i * max_length:(i + 1) * max_length] for i in range(num_chunks)]

    # Decode each chunk of token IDs back to text
    return [tokenizer.decode(chunk) for chunk in chunks]


# Process questions individually in chunks
results = []
for question in questions:
    chunks = chunk_text(question) # Chunk the question
    for chunk in chunks:
        result = teapot_ai.query(query="How many copies of Galactic Odyssey were sold?", context=chunk)
        results.append(result)

# Print or process the results as needed
for result in results:
    print(result)
```
